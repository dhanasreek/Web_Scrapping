## Web Scrapping : scrapping amazon reviews of a product

```python
import requests
from bs4 import BeautifulSoup
```

## Enter the url of the product for which we need to fetch the reviews.

```python
url=input("Enter the URL: ")
```

```python
url
```

```python
header={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}
```

```python
cookie={} # insert request cookies within{}
```

### A function to get the contents of individual product pages using 'data-asin' number (unique identification number)

```python
def Searchasin(asin):
    url="https://www.amazon.in/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

### A function to pass on the link of 'see all reviews' and extract the contentÂ¶

```python
def Searchreviews(review_link):
    url="https://www.amazon.in"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
data_asin = url.split("/dp/")[1]
data_asin =  data_asin.split("/")[0]
```

```python
data_asin = [data_asin]
```

```python
len(data_asin)
```

### By passing the data-asin numbers, we can extract the 'see all reviews' link for each product in the page

```python
link=[]
for i in range(len(data_asin)):
    response=Searchasin(data_asin[i])
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
pages=int(input("Enter Number of Pages to search : "))
```

```python
time=[]
for j in range(1):
    for k in range(pages):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-date"}):
            time.append(i.text)
```

```python
print(len(time))
print(time)
```

```python
names=[]
for j in range(1):
     for k in range(5):
            response=Searchreviews(link[j]+'&pageNumber='+str(k))
            soup=BeautifulSoup(response.content)
            for i in soup.find_all("span",class_='a-profile-name'):
                names.append(i.text)
```

```python
title=[]
for j in range(1):
    for k in range(pages):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("a",{'data-hook':"review-title"}):
            title.append(i.text)
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i])
review_title[:]=[titles.lstrip('\n') for titles in review_title]
review_title[:]=[titles.rstrip('\n') for titles in review_title]
review_title
```

```python
reviews=[]
for j in range(1):
    for k in range(pages):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            reviews.append(i.text)
```

```python
review=[]
for i in range(0,len(reviews)):
    review.append(reviews[i])
review[:]=[rev.lstrip('\n') for rev in review]
review[:]=[rev.rstrip('\n') for rev in review]
```

```python
print(len(review))
review
```

```python
ratings=[] 
for j in range(1):
    for k in range(pages):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("i",{'data-hook':"review-star-rating"}):
            ratings.append(i.text)
```

```python
rating=[]
for i in range(len(ratings)):
    rating.append(int(float(ratings[i].split(" ")[0])))
```

```python
len(rating)
```

```python
upvotes=[]
for j in range(1):
    for k in range(10):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"helpful-vote-statement"}):
            upvotes.append(i.text)
```

```python
len(upvotes)
```

```python
import pandas as pd
rev={'TimeStamp':time,'ReviewTitle':review_title,'review':reviews,'ratings':rating} #converting the reviews list into a dictionary
review_data=pd.DataFrame.from_dict(rev) #converting this dictionary into a dataframe
review_data.to_excel('Amazon_project_data.xlsx',index=False)
```
